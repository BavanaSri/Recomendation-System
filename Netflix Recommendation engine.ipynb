```python
# To find the unique number of users and movies in the 'ratings' dataset

n_users = df.Cust_Id.unique().shape[0]
n_movies = df.Movie_Id.unique().shape[0]
n_genre=df.Genre.unique().shape[0]
print(f'Number of users = {n_users} and Number of movies = {n_movies} and Number of Genres={n_genre}')
```

```python
# To create a rating matrix for the 'ratings' dataset

Ratings = df.pivot(index = 'Cust_Id', columns ='Movie_Id', values = 'Rating').fillna(0)
Ratings.head(10)
```

```python
# To install the scikit-surprise library for implementing SVD

from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate

# Load Reader library
reader = Reader()

# Load ratings dataset with Dataset library
data = Dataset.load_from_df(df[['Cust_Id', 'Movie_Id', 'Rating']], reader)

# Use the SVD algorithm.
svd = SVD()


# Compute the RMSE of the SVD algorithm.
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=3, verbose=True)
```

```python
# # Create a shallow copy for the movies dataset
#user_5 = df["Movie_Id","Genre","MovieName"].copy()
user_5 = df[['Movie_Id', 'Genre', 'MovieName']].copy()

# #Reset the index for user_5 dataset
user_5 = user_5.reset_index()



# # getting full dataset
data = Dataset.load_from_df(df[['Cust_Id', 'Movie_Id', 'Rating']], reader)


# #create a training set for svd
trainset = data.build_full_trainset()
svd.fit(trainset)

# #Predict the ratings for user5
user_5['Estimate_Score'] = user_5['Movie_Id'].apply(lambda x: svd.predict(1488844, x).est)

# #Drop extra columns from the user1 data frame
user_5 = user_5.drop(['Movie_Id','Genre','index'], axis = 1)

# # Sort predicted ratings for user1 in descending order
user_5 = user_5.sort_values('Estimate_Score', ascending=False)

# #Print top 10 recommendations
print(user_5.head(10))
```
